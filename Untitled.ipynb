## Step 1: Import Libraries

We'll import the essential libraries needed for data manipulation and visualization.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Better visuals
sns.set(style="darkgrid")
%matplotlib inline
```

## Step 2: Load the Dataset

We load the Netflix dataset (`netflix_titles.csv`) which contains details about movies and TV shows on the platform.


```python

df = pd.read_csv("netflix_titles.csv")
df.head()
```

## Step 3: Dataset Overview

Letâ€™s check how many rows and columns we have, and see the data types of each column.


```python
# Step 3: Basic Info
print("Shape of dataset:", df.shape)

# Show column names and data types
print("\nColumn Data Types:")
print(df.dtypes)

# Check for missing values
print("\nMissing Values in Each Column:")
print(df.isnull().sum())
```

## Step 4: Check for Missing Values

Weâ€™ll now check which columns have missing data so we can decide how to handle them.


```python
df.isnull().sum()
```

## Step 5: Data Cleaning

We'll clean the dataset to make it usable:
- Convert `date_added` to datetime format
- Fill missing values in `director`, `cast`, `country`
- Drop rows with missing `rating` or `duration` (only a few)


```python
# Fix date format: remove spaces and convert
df['date_added'] = pd.to_datetime(df['date_added'].str.strip(), errors='coerce')

# Fill missing values
df['director'].fillna('Unknown', inplace=True)
df['cast'].fillna('Unknown', inplace=True)
df['country'].fillna('Unknown', inplace=True)

# Drop few rows with critical missing values
df = df.dropna(subset=['rating', 'duration'])

# Recheck missing
df.isnull().sum()
```

```python
## Step 6.1: Type of Content on Netflix

Letâ€™s see how much of Netflixâ€™s content is `Movies` vs. `TV Shows`.
```

```python
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='type', palette='Set2')
plt.title('Distribution of Content Types on Netflix')
plt.xlabel('Type')
plt.ylabel('Count')
plt.show()
```

## Step 6.2: Content Over the Years

How has Netflixâ€™s content library grown over the years?
We'll count how many movies/shows were added per year.


```python
# Extract year from 'date_added'
df['year_added'] = df['date_added'].dt.year

plt.figure(figsize=(10,5))
sns.countplot(data=df, x='year_added', hue='type', palette='pastel')
plt.title('Content Added to Netflix Over the Years')
plt.xticks(rotation=45)
plt.xlabel('Year Added')
plt.ylabel('Number of Titles')
plt.legend(title='Type')
plt.show()
```

## Step 6.3: Top 10 Countries by Content

Which countries have produced the most Netflix content?
Weâ€™ll look at the top 10.


```python
# Count countries (first listed country only)
top_countries = df['country'].str.split(',').str[0].value_counts().head(10)

plt.figure(figsize=(10,5))
sns.barplot(x=top_countries.values, y=top_countries.index, palette='viridis')
plt.title('Top 10 Countries with Most Content on Netflix')
plt.xlabel('Number of Titles')
plt.ylabel('Country')
plt.show()
```

## Step 6.4: Most Common Genres

Which genres (categories) appear most frequently?
We'll use the `listed_in` column.


```python
# Split genres and count
from collections import Counter

genres = df['listed_in'].str.split(', ')
genre_counts = Counter([genre for sublist in genres for genre in sublist])
top_genres = genre_counts.most_common(10)

# Plot
plt.figure(figsize=(10,5))
sns.barplot(x=[x[1] for x in top_genres], y=[x[0] for x in top_genres], palette='coolwarm')
plt.title('Top 10 Genres on Netflix')
plt.xlabel('Number of Titles')
plt.ylabel('Genre')
plt.show()
```

## Step 6.5: Movie Duration Distribution

How long are Netflix movies? Letâ€™s analyze the `duration` column for movies only.


```python
# Filter movies and convert duration to minutes
movies_df = df[df['type'] == 'Movie'].copy()
movies_df['duration_minutes'] = movies_df['duration'].str.extract('(\d+)').astype(float)

plt.figure(figsize=(10,5))
sns.histplot(movies_df['duration_minutes'], bins=30, kde=True, color='coral')
plt.title('Distribution of Movie Durations on Netflix')
plt.xlabel('Duration (Minutes)')
plt.ylabel('Frequency')
plt.show()
```

## ðŸ“Œ Conclusion & Key Insights

After analyzing the Netflix dataset, we found several interesting insights:

- **Content Type**: Netflix has more movies than TV shows, but both are significant.
- **Growth Over Time**: The number of titles has increased steadily, especially from 2015 to 2020.
- **Country Distribution**: The **United States** dominates Netflix content, followed by **India**, **United Kingdom**, and others.
- **Top Genres**: Popular genres include *Dramas*, *Comedies*, *Documentaries*, and *Children's programming*.
- **Movie Length**: Most Netflix movies are around **90â€“120 minutes** in duration.
